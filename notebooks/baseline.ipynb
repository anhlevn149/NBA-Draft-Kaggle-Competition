# 2. Data Preparation

## 2a. Training Dataset

```python
# Make a copy of df_train and save as df_train_cleaned
df_train_cleaned = df_train.copy()
```

First, we are going to remove columns whose datatype is object.

```python
# Remove columns with objects datatype from the dataset
df_train_cleaned.drop(df_train[['team', 'conf', 'yr', 'ht', 'num', 'type', 'player_id']], axis=1, inplace=True)
```

```python
# Check if there is any columns with missing values
df_train_cleaned.isna().sum()
```

For those columns with missing values more than 50% of the total values, we decide to remove them from the dataframe. For the rest, we will fill the missing values with the mean value of that column. 

```python
# Remove columns with more than 50% missing values from the dataframe
df_train_cleaned.drop(df_train[['Rec_Rank', 'dunks_ratio', 'pick']], axis=1, inplace=True)
```

```python
# Replace all missing values with mean values of corresponding columns
df_train_cleaned = df_train_cleaned.fillna(df_train_cleaned.mean())
```

```python
# Check if there is still any missing values
df_train_cleaned.isna().sum()
```

So all the rows have been filled with values. We will take a look at the cleaned dataframe before pre-processing it.

```python
# Show the first 5 rows of the df_train_cleaned dataframe
df_train_cleaned.head()
```

## 2c. Set features and target variables

```python
# Drop column 'drafted' from df_train_cleaned and save as an variable called X_data
X_data = df_train_cleaned.drop('drafted', axis='columns')
```

```python
# Save name of the target column as a variable called y_data
y_data = df_train_cleaned['drafted']
```

```python
# Print the shape of X_data and y_data
print(X_data.shape)
print(y_data.shape)
```

```python
# Assign df_test_cleaned as an variable called X_test
X_test = df_test_cleaned
```

# 3. Split Dataset

```python
# Import the train_test_split package from scikit-learn
from sklearn.model_selection import train_test_split
```

```python
# Split the training dataset to training data (80%) and validation data (20%)
X_train, X_val, y_train, y_val = train_test_split(X_data, y_data, test_size=0.2, random_state=42, stratify=y_data)
```

```python
# Print the shape of X_train, X_val and X_test
print(X_train.shape)
print(X_val.shape)
print(X_test.shape)
```

```python
# Print the shape of y_train, y_val
print(y_train.shape)
print(y_val.shape)
```

# 4. Scale Data

```python
# Import StandardScaler from scikit-learn to scale the data
from sklearn.preprocessing import StandardScaler
```

```python
# Instantiate the StandardScaler method and save as a variable called scaler
scaler = StandardScaler()
```

```python
# Fit the training dataset to the scaler
scaler.fit(X_train)
```

```python
# Apply the scaler to the training, validation and testing datasets
X_train = scaler.transform(X_train)
X_val = scaler.transform(X_val)
X_test = scaler.transform(X_test)
```

```python
# Import dump from joblib
from joblib import dump
```

```python
# Save scaler into the folder models and call the file as scaler.joblib
dump(scaler, '../models/scaler.joblib')
```

```python
# Save the sets to folder data/processed
pd.DataFrame(X_train).to_csv('../data/processed/X_train.csv', index=False)
pd.DataFrame(X_val).to_csv('../data/processed/X_val.csv', index=False)
pd.DataFrame(X_test).to_csv('../data/processed/X_test.csv', index=False)
pd.DataFrame(y_train).to_csv('../data/processed/y_train.csv', index=False)
pd.DataFrame(y_val).to_csv('../data/processed/y_val.csv', index=False)
```
